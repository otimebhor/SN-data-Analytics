1. str.capitalize() : Converts the first character to upper case
Application: It is can be used to convert 'book' to 'Book'
Code block:
string = 'book
string = string.capitalize()
print(string)

2. str.swapcase() : Returns a copy of the string with all its uppercase letters converted into lower case and vice versa.
Application: It can convert "I am a Girl' to i AM A gIRL'
Code block:
string = 'I am a Girl'
string = string.swapcase()
print(string)

3. str.startswith() : Returns True if the string starts with the specified value; otherwise, it returns False.
Application: It can be used to check if a word was used at the beginning of a sentence.
Code block:
string = 'Maybe I can'
string = string.startswith('Maybe')
print(string)

4. str.endswith() : Returns True if the string ends with the specified value, else it returns False.
Application: It can be used to check if a word was used at the end of a sentence.
Code block:
string = ''There is nobody at home'
string = string.endswith('nobody')
print(string)

6. str.upper() : Capitalizes all the letters of the string.
Application: It can be used to convert 'cattle' to 'CATTLE'
Code block:
string = 'cattle'
string = string.upper()
print(string)

7. str.title() : Capitalizes all the first letters of the given string.
Application: It can be used to convert 'lagos state' to 'Lagos State'.
Code block:
string = 'lagos state'
string = string.title()
print(string)

8. str.lower() : Returns all the letters of the string to lowercase letters.
Application: It can be used to convert 'Boat Cruise' to 'boat cruise'
Code block:
string = 'Boat Cruise'
string = string.lower()
print(string)

9. str.replace() : Takes an original string and return an updated string with some replacement.
Application: It can be used to replace 'has' to 'had' in a statement.
Code block: 
string = 'Sammy has a ballon'
string = string.replace('has', 'had')
print(string)

10. str.isnumeric() : Returns True if the string starts with the specified value; otherwise, it returns False.
Application: Checks if 'house' consists of only numeric characters.
Code block:
string = 'house'
print(string.isnumeric())


11. str.isupper() :  Returns True if the string's alphabetic characters are all uppercase ; otherwise, it returns False.
Application: It can be used to check if 'A SAMMY ODYSSEY' is made up of only uppercase characters.
Code block:
string = 'A SAMMY ODYSSEY'
print(string.isupper())

12. str.islower() : Returns True if the string's alphabetic characters are all lowercase; otherwise, it returns False.
Application: It can be used to check if 'we lived in a small town' is made up of only lowerercase characters.
Code block:
string = 'we lived in a small town'
print(string.islower())

13. str.istitle() : Returns True if the string is in title case; otherwise, it returns False.
Application: It can be used to check if 'A Thousand Splendid Sharks' is in title case. 
Code block:
string = 'A Thousand Splendid Sharks'
print(string.istitle())

14. str.find() : Searches the string for a specified value and returns the position of where it is found.
Application: It can be used to know if 'car' is in 'I own a black car and a big house' and its position.
Code block:
string = 'I own a black car and a big house'
string = string.find('car')
print(string)

15. str.index() : Searches the string for a specified value and returns the position of where it is found.
Application: It can be used to know if 'r' is in 'scared' and its position.
Code block:
string = 'scared'
string = string.index('r')
print(string)
